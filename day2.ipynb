**operator**
1. Arithematic + , - , / , * , % , // , **

2. relational < , > , <= , >= , == ,  !=
3. assignment = , += , -= , *= , /= , %=

4. logical and , or , not
5. bitwise & , |

```python

# Arithematic
num1 = 10
num2 = 20

print(num1 + num2)
print(num1 - num2)
print(num1 * num2)
print(num1 / num2)
print(num1 % num2)
print(num1 // num2)
print(num1 ** num2)
```

```python

# Relational

num1 = 10
num2 = 20

print(num1 < num2)
print(num1 > num2)
print(num1 <= num2)
print(num1 >= num2)
print(num1 == num2)
print(num1 != num2)
```

```python
a = num1 << 2
print(a)
```

```python
# Assignment

num1 = 10
num2 = 20

num1+=num2
print(num1)  # num1 = 30

num1-=num2
print(num1)

num1*=num2
print(num1)

num1/=num2
print(num1)

num1%=num2
print(num1)
```

```python

# Logical imp

num1 = 10
num2 = 20


# and
print(num1 and num2)   # T1 and T2 = T2 , if 1st true then we check 2nd if 2nd also true  , the return value is 2nd

print(num1 and not num2) # T and F = F  ans is false

print(not num1 and num2) # F and T = F

print(not num1 and not num2) # F and F = F


print([] and num1)
print(num1 and 0)
```
