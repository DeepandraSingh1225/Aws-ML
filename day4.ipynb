```python

lst = [1,2,3,4,5]
x = 100
i = 1

lst.append(x)      # Add item at end
lst.extend([1,2,3])   # Add multiple items
lst.insert(i, x)   # Insert at index

lst.remove(x)      # Remove first occurrence
lst.pop(i)         # Remove and return item (default last)
lst.clear()        # Remove all items

lst = [1,2,3,4,5]
lst.index(4)       # Find index of item
lst.count(x)       # Count occurrences

lst.sort()         # Sort list in-place
lst.reverse()      # Reverse list in-place

lst.copy()         # Shallow copy
```

```python

d = {"a" : 1,
     "b" : 2}


d.clear()          # Remove all items
print(d)
d.copy()           # Shallow copy

d = {"a" : 1,
     "b" : 2}

print(d.get("a"))         # Get value safely
d.setdefault("c" , 500) # Return value, insert if absent


print(d)

print(d.keys())           # View keys
print(d.values())         # View values
print(d.items())          # View key-value pairs


d1 = {"a" : 1,
      "b" : 2}
d2 = {"c" : 3,
      "d" : 4}

d1.update(d2)      # Merge dictionaries

d.update(d1)    # Merge dictionary

d.pop("a")         # Remove and return value
d.popitem()        # Remove last inserted pair
```
