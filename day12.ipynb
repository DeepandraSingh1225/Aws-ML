```python
import pandas as pd

l = [1,2,3,4,5]
df = pd.Series(l)              # convert lis into series
print(df)
```

```python
df = pd.Series(data = l)
print(df)
```

```python
df = pd.Series(data = l, index = ['a','b','c','d','e'])
print(df)
```

```python
d = {'a':1,'b':2,'c':3,'d':4,'e':5}
df = pd.Series(d)
print(df)
print()

print(f"Element at index 'a' -> {df['a']}")
```

```python
l = [1,2,3,4,5]
df = pd.Series(l , copy = False)
```

```python

data = {
    'name' : ['nikhil' , 'raj' , 'dev' , 'hello'],
    'age' : [10,20,30,40]
}

# 1d
df = pd.Series(data)
print(df)

# 2d
df = pd.DataFrame(data)
print(df)
```

```python
data = [['nikhil' , 10] , ['raj' , 20] , ['dev' , 30] , ['hello' , 40]]
df = pd.DataFrame(data)
print(df , '\n \n')

# Add column name
df = pd.DataFrame(data , columns = ['name' , 'age'] )
print(df , '\n \n')


#Add index name
df = pd.DataFrame(data , columns = ['name' , 'age'] , index = ['a','b','c','d'])
print(df )
```
