```python
# Arithematic operator

num1 = 10
num2 = 20

print(num1 + num2)
print(num1 - num2)
print(num1 * num2)
print(num1 / num2)
print(num1 % num2)
print(num1 // num2)            # floor division
print(num1 ** num2)            # power
```
