```python
# Arithematic operator

num1 = 10
num2 = 20

print(num1 + num2)
print(num1 - num2)
print(num1 * num2)
print(num1 / num2)
print(num1 % num2)
print(num1 // num2)            # floor division
print(num1 ** num2)            # power
```

```python
#Truthy and Falsy values
# all num except 0 is true
a = 0
print(not a)


# all string except empty string is true
str = ''
print(not str)

# empty list , tuple , dict , set  is false
```

```python
# And
a = 10
b = 20

# T and T = T
print(a and b) # ans is 20 because it check 2nd value

# T and F = F
print(a and not b) # check 2nd value because 1st is true ans is false because Not convert value 20 into false

# F and T = F
print(not a and b) # check 1st value because 1st is false ans is false

# F and F = F
print(not a and not b)
```

```python
a = 10
b = 20

print( a or b)

# because a is true is dont check value of b so ans is 10
  # T or T = T

  # T or F = T
print( a or b)   # same case as above
  # F or T = T
print(not a or b)  # check 2nd value because 1st is false ans is b

  # F or F = F
print(not a or not b)  # check 2nd value because 1st is false ans is false because not convert value to false
# similarly
```
