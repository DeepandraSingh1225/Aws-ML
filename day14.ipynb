#### Handle missing values
- either drop null values
- fill with avg
-

```python
import numpy as np
import pandas as pd

data = {
    'experience' : [1,np.nan,3,4,5],
    'salary' : [100,200,300,400,np.nan],
    'name' : ['hello' , 'dheeraj' , 'Nikhil' , 'Raj' , 'vinay']
}

df = pd.DataFrame(data)
df
```

```python
# null value filter
print(df.isnull() , '\n\n')

# sum of null | to find num of null value
print(df.isnull().sum() , '\n\n')

# dropna | drop complete row
df.dropna()
print(df)

# fillna | fill any value
df.fillna(5)
print(df)
```
