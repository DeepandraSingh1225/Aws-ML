```python
def calculator( a , b , sign):
  if sign == '+':
    print('addition' , a+b)
  elif sign == '-':
    print('sub' , a-b)
  elif sign == '*':
    print('mul' , a*b)
  elif sign == '/':
    print('division' , a/b)

num1 = int(input("Enter a num"))
num2 = int(input("Enter a num"))
sign = input("Enter sign")
calculator(num1 , num2 , sign)
```

## Exception handling
  - ithas TRY , Except , Final

```python
try:
  print("try part")
except:
  print("Except part")
```

```python
try:
  print("try part")
  print("hello"+10)
except:
  print("Except part")
```

```python
try:
  print(10/0)
except TypeError:
  print("type error")
except ZeroDivisionError:
  print("zero division error")
```

```python
try:
  print("try part")
except:
  print("Except part")
finally:
  print('finally part')
```

```python
try:
  print(10/0)
except:
  print("Except part")
finally:
  print('finally part')
```
