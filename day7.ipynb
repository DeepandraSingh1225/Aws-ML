### Inheritance

```python
class User1:
  def __init__(self,pen,pencil):
    print("constructor of user1")
    self.pen = pen
    self.pencil = pencil
  def penGet(self):
    print("pen is working: ",self.pen)
  def pencilGet(self):
    print("pencil is working: ",self.pencil)



class User2(User1):
  def __init__(self,pen,pencil,earser):
    print("constructor of user2")
    self.pen = pen
    self.pencil = pencil
    self.earser = earser
# object
u = User2("blue","black","doms")
u.penGet()
```

```python
class nikhil:
  def __init__(self,bike):
    self.bike = bike

  def usebike(self):
    print("bike used by: ",self.bike)

class rajendra(nikhil):
  def __init__(self,name,bike):
    super().__init__(bike)
    self.name = name

  def usecar(self):
    print("car used by: ",self.name)

# object instantiation and method calls are outside the class definition
u = rajendra("rajendra jangid","tvs by rajendra")
u.usecar() # Corrected from u.usename()
u.usebike()
```

### Encapsulation


```python
# encapsulation
class Student:
  def __init__(self,name,roll_no):
    self.__data = name # name is attribute
    self.roll_no = roll_no # roll_no is attribute
  def getname(self,password=None):
    if password == "123456":
      print(self.__data)
    else:
      print("password not match")

s = Student("rajendra",123)
# print(s.__data) # no attribute
print(s.roll_no)
s.getname("123456")
```

```python
class Phone:
    def __init__(self, pin):
        self.__pin = pin

    def unlock(self, entered_pin):
        if entered_pin == self.__pin:
            print("Phone Unlocked")
        else:
            print("Wrong PIN")

phone = Phone("4321")

phone.unlock("4321")
phone.unlock("1111")

```

### Class variable

```python
class Student:
  count = 0 # count is class variable
  def __init__(self , name ,roll):
    print("Constructor")
    self.name = name
    self.roll = roll
    Student.count += 1

  def getname(self):
    print(self.name)



s = Student("rajendra",123)
s.getname()
print(s.count)
```

- overriding

```python
class User1:
  def __init__(self,pen,pencil):
    self.pen = pen
    self.pencil = pencil
  def getpen(self):
    print("user1: ",self.pen)



class User2(User1):
  def __init__(self,pen,pencil,earser):
    self.earser = earser
    super().__init__(pen,pencil)
  def getpen(self):
    print("user2: ",self.pen)

u = User2("blue","black","doms")
u.getpen()
```
