- we can not make 2d array using arange so we use reshape

```python
arr = np.arange(6)
print(arr)
print()

arr1 = arr.reshape(2,3)
print(arr1)
```

#### flatten

```python
arr1 = arr.reshape(2,3)
print(arr1)
print()

arr1.flatten()
```

#### Ravel

```python
arr = np.arange(6).reshape(2,3)
print(arr)

arr.ravel()
```

#### flatten vs ravel


```python
print("Flatten")
arr = np.arange(4).reshape(2,2)
print(arr)
print()

arr_new = arr.flatten()
arr_new[0] = 100
print("New :-" , arr_new)
print("Original:-")
print(arr)
print()

print("Ravel")
arr1 = np.arange(4).reshape(2,2)
print(arr1)
print()

arr_new1 = arr1.ravel()
arr_new1[0] = 200
print("New :-" , arr_new1)
print("Original:-")
print( arr1)
```

#### Transpose

```python
arr = np.arange(6).reshape(2,3)
print(arr)
print()

arr1 = arr.T
print(arr1)
print()

print(arr.transpose())
```
