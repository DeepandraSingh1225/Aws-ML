# Class

```python
class student:
  def __init__(self , name , age , grade):
    self.name = name
    self.age = age
    self.grade = grade

s = student('Deepandra' , 21 , 'A')

print(s.name)
print(type(s.name))
print(s.age)
print(type(s.age))
print(s.grade)
```

```python
# we can use any word in place of self keyword

class student:
  def __init__(s , name , age , grade):
    s.name = name
    s.age = age
    s.grade = grade

s = student('Deepandra' , 21 , 'A')
```

```python
class Student:
  def __init__(self , name , age , roll):
    self.name = name
    self.age = age
    self.roll = roll

# sequence can be anything
s = Student(age = 21 , name = 'Deepandra' , roll = 10)

print(s.name)
print(s.age)
print(s.roll)

```
