```python
class student:
  def __init__(self , name , age , grade):
    self.name = name
    self.age = age
    self.grade = grade

s = student('Deepandra' , 21 , 'A')

print(s.name)
print(type(s.name))
print(s.age)
print(type(s.age))
print(s.grade)
```

```python
class Student:
  def __init__(self , name = None, age=None , grade=None):
    self.name = name
    self.age = age
    self.grade = grade

s = Student('Deepandra')
print(s.name , s.age , s.grade)

s1 = Student('Deepandra' , 21)
print(s1.name , s1.age , s1.grade)

s2 = Student('Deepandra' , grade = 'A')
print(s2.name , s2.age , s2.grade)

s = Student()
print(s.name , s.age , s.grade)
```

## Inheritance

```python
class Student:
  def __init__(self,name , roll):
    self.name = name
    self.roll = roll

  def getData(self):
    print(self.name , self.roll)


#inheritance
class Address(Student):
  def __init__(self ,name , roll , city , state):
    self.city = city
    self.state = state
  #extra
    self.name = name
    self.roll = roll


  def getAddress(self):
    print(self.city , self.state)




# s = Student('deepandra' , 123)
# s.getData()

a = Address('deepandra' , 123,'jaipur' , 'rajasthan')
a.getAddress()
a.getData()
```

```python
# we repeat this part in both classes
  # self.roll = roll
  # self.name = name



class Student:
  def __init__(self,name , roll):
    self.name = name
    self.roll = roll

  def getData(self):
    print(self.name , self.roll)


#inheritance
class Address(Student):
  def __init__(self ,name , roll , city , state):
    self.city = city
    self.state = state
    super().__init__(name , roll)


  def getAddress(self):
    print(self.city , self.state)




# s = Student('deepandra' , 123)
# s.getData()

a = Address('deepandra' , 123,'jaipur' , 'rajasthan')
a.getAddress()
a.getData()
```
