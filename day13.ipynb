#### Operation on col


```python
import pandas as pd

data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}

df = pd.DataFrame(data)

# Single column
print(df['Name'] , '\n\n')

# col value change
df['Emp_ID'] = [1,2,3,4,5,6]
print(df['Emp_ID']  , '\n\n')


# Add new col
df['Amount'] = df['Salary']
print(df['Amount']  , '\n\n')

#Add new cols manually
df['Increment'] = [0,0,0,0,0,0]
print(df , '\n\n')

# multiple column
print(df[['Name' , 'Salary']]  , '\n\n')

df
```

```python
# drop col
df.drop(columns = ['Increment' , 'Amount'] , axis=1, inplace = True)
print(df)
```

#### Operation on row


```python
import pandas as pd

data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}

df = pd.DataFrame(data)

print(df , '\n\n')

# loc    (row-> idx , col-> col name)
print(df.loc[1,'Name'] , '\n\n')

# iloc   (row -> idx , col-> idx)
print(df.iloc[1,1] , '\n\n')

# Multiple rows
print(df.loc[0:3], '\n\n' )

print(df.loc[[0,1,2]], '\n\n')

print(df.iloc[[1,2,3]], '\n\n')
```

```python
print(df.iloc[[0,1] , [0,1]])
```

#### Filter

```python
import pandas as pd

data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}

df = pd.DataFrame(data)

print(df , '\n\n')

# emp_Id = 106  , Id >104
print(df[df['Emp_ID'] > 104] , '\n\n')
print(df[df['Emp_ID'] == 106])

# filter in salary
print(df[(df['Salary'] >30000) & (df['Salary']) < 50000])
```
