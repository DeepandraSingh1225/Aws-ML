#### Filter

```python
import numpy as np
arr = np.array([10,20,3,4,90,200])
arr_f = arr[arr > 20]
arr_f
```

```python
arr = np.array([10,20,11,12,13,14,15,16,29,3,4,90,200])
arr_f = arr[(arr < 20) & (arr>10)]
arr_f
```

#### Where

```python
arr = np.array([10,20,30,40])
print(np.where(arr > 30 , 'true' , 'false'))

print(np.where(arr > 20 , arr*2 , arr))
```

#### Fancy indexing

```python
arr = np.array([10,20,3,4,90,200])
print(arr)
arr_f = arr[[0,2]]
print(arr_f)
```
